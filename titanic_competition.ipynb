```python
#encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

titanic['Sex'] = encoder.fit_transform(titanic['Sex'])
titanic['Embarked'] = encoder.fit_transform(titanic['Embarked'])
titanic['Cabin'] = encoder.fit_transform(titanic['Cabin'])
titanic['Ticket'] = encoder.fit_transform(titanic['Ticket'])
```

```python
#ip op creation
x = titanic.drop('Survived',axis=1)
y=titanic['Survived']
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
#standard scaler transform
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
#model training
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train , y_train)
```

```python
#prediction of training data
training_data_prediction = lr.predict(x_train)
```

```python
#evaluation of training data
from sklearn.metrics import r2_score
r2_score = r2_score(y_train , training_data_prediction)
print(r2_score)
```

```python
#prediction of test data
test_data_prediction = lr.predict(x_test)
```

```python
#evaluation of test data
from sklearn.metrics import r2_score
r2 = r2_score(y_test , test_data_prediction)
print(r2)
```
